5. DNA sequence analyzer

```python
sequence="ATTGCATGCGTACAGATGCATCGAATGCATGCATGATTTTTGCGCATGCATGCATGCCTC"
print("sequnece length:",len(sequence))
a=sequence.count("A")
b=sequence.count("T")
c=sequence.count("C")
d=sequence.count("G")
print("A:","T:","C:","G:",a,b,c,d)
```

☹ example for DNA sequence analyzer

```python
sequence="ATTGCATGCGTACAGATGCATCGAATGCATGCATGATTTTTGCGCATGCATGCATGCCTC"
print("sequnece length:",len(sequence))
a=sequence.count("A")
t=sequence.count("T")
g=sequence.count("C")
c=sequence.count("G")
print("A:","T:","C:","G:",a,t,g,c)
gc_content=((g +c)/len(sequence))*100
print("GC content:",round(gc_content,2),"%")

if gc_content>50:
  print("high gc gene - stable structure")
else:
  print("low gc gene - AT-rich region")
```
